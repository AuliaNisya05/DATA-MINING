```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Baca file dan tampilkan seluruh data
file_name = '/content/drive/MyDrive/Colab Notebooks/Praktikum 1/Copy of 500_Person_Gender_Height_Weight_Index.csv'
data = pd.read_csv(file_name)
print("Seluruh Data:")
print(data)

# 2. Gunakan fungsi statistik untuk menghitung jumlah data, rata-rata,nilai maksimum, nilai minimum, dari berat dan tinggi badan berdasarkan Gender
def statistik_data(data):
    statistik = data.groupby('Gender')[['Weight', 'Height']].agg({
        'Weight': ['count', 'mean', 'std', 'min', 'max'],
        'Height': ['count', 'mean', 'std', 'min', 'max']
    })
    return statistik

statistik = statistik_data(data)
print("\nStatistik Berat dan Tinggi Badan berdasarkan Gender:")
print(statistik)

# 3. Ubah kolom gender menjadi data binary: Female – 0, Male – 1
data['Gender_Binary'] = data['Gender'].map({'Female': 0, 'Male': 1})

# 4. Tampilkan grafik Index BMI berdasarkan gender
plt.figure(figsize=(10, 6))
plt.bar(data['Gender'], data['Index'])
plt.xlabel('Gender')
plt.ylabel('Index BMI')
plt.title('Grafik Index BMI berdasarkan Gender')
plt.show()
```
